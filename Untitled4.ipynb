```python
from bs4 import BeautifulSoup

html = '''
<div class="fMghEO">
    <ul class="_1xgFaf">
        <li class="rgWa7D">12 GB RAM | 256 GB ROM</li>
        <li class="rgWa7D">17.02 cm (6.7 inch) Full HD+ Display</li>
        <li class="rgWa7D">50MP (OIS) + 50MP | 32MP Front Camera</li>
        <li class="rgWa7D">5000 mAh Battery</li>
        <li class="rgWa7D">Dimensity 7200 Pro Processor</li>
        <li class="rgWa7D">1 Year Manufacturing Warranty</li>
    </ul>
</div>
'''

soup = BeautifulSoup(html, 'html.parser')

# Extracting specifications and categorizing them
spec_categories = {
    'Storage': ['GB ROM'],
    'Display': ['inch'],
    'Camera': ['MP', 'OIS'],
    'Battery': ['mAh'],
    'Processor': ['Processor'],
    'Warranty': ['Warranty']
}

specs = {}
for category, keywords in spec_categories.items():
    specs[category] = [li.text.strip() for li in soup.find("ul", class_="_1xgFaf").find_all("li", class_="rgWa7D") if any(keyword in li.text for keyword in keywords)]

# Printing categorized specifications
for category, spec_list in specs.items():
    print(f"{category}:")
    for spec in spec_list:
        print("-", spec)
```
